# CEOSAL

## Data

```python
%%capture

!pip install wooldridge
!pip install stargazer

import wooldridge as woo
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from stargazer.stargazer import Stargazer

data = woo.data('ceosal2')
```

## Plot

```python
x = data['ceoten']
y = data['salary']

plt.scatter(x, y)
plt.show()
```

# Regression

```python
model = smf.ols('salary ~ I(ceoten**2)', data)

results = model.fit()

print(results.summary())
```

## Renaming the covariants

```python
stargazer = Stargazer([results])
stargazer.rename_covariates({'I(ceoten ** 2)': 'ceoten^2'})
stargazer
```

# Customizing plots
Matplotlib plots are highly customizable. You can change the labels, legends, colors, and markers to make your plots more informative and visually appealing.

### Figure size

```python
# Create a figure
plt.figure(figsize=(10, 8))

# Create x and y variables
x = data['ceoten']
y = data['salary']

# Create a line plot
plt.scatter(x, y)
# Display the plot
plt.show()
```

### Adding labels and title

```python
plt.scatter(x, y)

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Color

```python
plt.scatter(x,
            y,
            color='green')

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Marker

```python
plt.scatter(x,
            y,
            color='green',
            marker='*')

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Size

```python
plt.scatter(x,
            y,
            color='green',
            marker='*',
            s=100)

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Transparency

```python
plt.scatter(x,
            y,
            color='green',
            marker='*',
            s=100,
            alpha=0.1)

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Border

```python
plt.scatter(x,
            y,
            color='green',
            marker='*',
            s=100,
            alpha=0.5,
            edgecolors='black',
            linewidths=1)

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Adding a legend

```python
plt.scatter(x,
            y,
            color='green',
            marker='*',
            s=100,
            alpha=0.5,
            edgecolors='black',
            linewidths=0.5,
            label='Salary')

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

# Adding the legend
plt.legend()

plt.show()
```

## Multiple plots

```python
z = data['mktval']

plt.scatter(x, y, label='Salary')

# Adding a second plot
plt.scatter(x, z, label='Market Value', color='red', alpha=0.5)

plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Salary and Market Value vs. Year')
plt.legend()

plt.show()
```

## Subplot

```python
# Create first subplot
plt.subplot(2, 1, 1)
plt.scatter(x, y, label='Salary')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Salary vs. Year')
plt.legend()

# Create second subplot
plt.subplot(2, 1, 2)
plt.scatter(x, z, label='Market Value', color='red')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Market Value vs. Year')
plt.legend()

# Adjust layout
plt.tight_layout()

plt.show()
```
